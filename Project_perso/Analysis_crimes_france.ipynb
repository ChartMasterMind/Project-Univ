```python

    
for i in crime["annees"].unique():


        crime_homicide_filtre = crime.loc[(crime["statut"] == "Homicides" ) & (crime["annees"] == i )]

      
        bar = plt.bar( i ,crime_homicide_filtre["faits"].sum(), color = "#16e5cb")

        plt.title(f"Les homicides en france en {i}")

        def ajout_etiquette_barre(ax, bars):

            for bar in bars: 

                yvaleur = bar.get_height()  
                ax.text(bar.get_x() + bar.get_width() / 2 , yvaleur + 0.9, round(yvaleur, 1), ha='center', va='bottom', fontsize=8)

        ajout_etiquette_barre(plt.gca(), bar)

    
```

```python
# je vais automatiser 

for i in range (2016,2024):

    
    filtre = crime.loc[(crime['statut'] == 'Homicides') & (crime['annees'] == i)] 

    filtre = filtre.sort_values(by = "faits" , ascending =False) 
    plt.figure(figsize=(15,6)) 

    bars = plt.bar(filtre["region"], filtre["faits"])
    

    plt.xticks(rotation=45, ha='right')  


    
    def colorier_le_max_et_le_min (a, b , color_max=str , color_min=str): 
        for bar in a :  
            if bar.get_height() == max (b) :
                bar.set_color(color_max)
            elif bar.get_height() == min(b):
                bar.set_color(color_min)


    colorier_le_max_et_le_min(bars, filtre["faits"] ,"red" , "green") 
   
    # Fonction pour ajouter les nombre d'homicide au-dessus des barres

    # Ajouter les valeurs au-dessus des barres
    ajout_etiquette_barre(plt.gca(), bars) 

    plt.title(f"Le nombre d'homicide dans les différentes régions francaises en {i}" )
```

```python
# moyenne des homicide entre 2016 et 2023


moyennes = []

for crimes_statut in crime['statut'].unique():

    for region in crime['region'].unique():

        crime_filtre = crime[(crime['statut'] == crimes_statut) & (crime['region'] == region)]

        Faits = crime_filtre["faits"].sum()


        moyennes.append({'statut' : crimes_statut , "region" : region , "fait" : Faits })



        moyennes_df = pd.DataFrame(moyennes)

# crée un graphique pour chaque statut 

for crimes_statut in moyennes_df["statut"].unique():

    moyenne_par_statut = moyennes_df[moyennes_df['statut'] == crimes_statut]

    moyenne_par_statut = moyenne_par_statut.sort_values(by = "fait" , ascending = False) 


    # affichage du graphique 

    plt.figure(figsize=(15, 6))
    
    def colorier_le_max_et_le_min_float (a,b, color_max= str , color_min =str): # car l'autre focntion ne marchait pas aveec les float, le a designe la colonne ou se trouve les float et b rpz les barres du grpahique
        max_height = max(a)
        min_height = min(a)
        # cette boucle va faire ca : on a un tuple ( barre , a) cad la premiere iteration sera ( barre 1 du grpahique , 26.6), si 26.6 est le max de la colonne moyenne faits, alors la barre 1 sera colorier en rouge sinon en vert si c'est le min, ou en bleu si c'est aucun des deux ( bleu est par default)
        for b, a in zip(b, a): # en gros cette  bloucle va caluler pour chaque b ( la barre) le a maximal et minimal. pour cela on fait un tuple pour comparer la barre et le a (= la moyenne) associé
            if a == max_height:
                b.set_color(color_max)
            elif a == min_height:
                b.set_color(color_min)
    
    bares2 = plt.bar (moyenne_par_statut['region'], moyenne_par_statut['fait'], color ="grey") 
    plt.xticks(rotation = 45 , ha = 'right')

    plt.title( f"Total des {crimes_statut} commis entre 2016 et 2023 dans les régions francaises ")
    
    # ajout des fonction 
    
    colorier_le_max_et_le_min_float(moyenne_par_statut['fait'], bares2, color_max ="blue", color_min ="yellow")

    ajout_etiquette_barre(plt.gca(),bares2)


        
```

```python
# population moyenne


moyenne_population = crime.groupby("region")["nombre_population"].mean() 

# graphique en barre avec pandas pour l'integrer avec groupe by 
plt.figure(figsize=(10, 6))

moyenne_population.plot(kind='bar')




```

```python



moyenne_population = crime.groupby("region")["nombre_population"].mean().sort_values(ascending=False)
plt.figure(figsize=(20,6))

bares3 = plt.bar(moyenne_population.index, moyenne_population.values, color ='#16e5cb')


plt.xticks(rotation = 45 , ha= "right")

ajout_etiquette_barre(plt.gca(), bares3)

colorier_le_max_et_le_min_float(moyenne_population , bares3 , color_max= '#7644e3' , color_min= "yellow")

plt.title(" Population moyenne des diférentes regions francaises entre 2016 et 2023")
```

```python
moyenne_crime_totaux = crime.groupby("region")["faits"].mean().sort_values(ascending=False)

plt.figure(figsize=(20,10))

bares4 = plt.bar(moyenne_crime_totaux.index, moyenne_crime_totaux.values , color = "grey")

plt.xticks(rotation = 45 , ha = "right")

plt.title("Moyenne des crime commis dans les différentes régions entre 2016 et 2023")


colorier_le_max_et_le_min_float(moyenne_crime_totaux, bares4 , color_max="red" , color_min="green")

ajout_etiquette_barre(plt.gca(), bares4)
```
